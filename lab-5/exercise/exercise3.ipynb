### Analyse the convergence properties of the Jacobi and Gauss-Seidel methods for the solution of a linear system whose matrix is
### $$\left[\begin{matrix}
    \alpha &&0 &&1\\
    0 &&\alpha &&0\\
    1 &&0 &&\alpha
    \end{matrix}\right],
    \quad \quad
    \alpha \in \mathbb{R}.$$

Let $$ A = \left[\begin{matrix}
    \alpha &&0 &&1\\
    0 &&\alpha &&0\\
    1 &&0 &&\alpha
    \end{matrix}\right],
    \quad \quad
    \alpha \in \mathbb{R}.$$
<br><br>
Write $A = D+L+U$, where $D$ is the diagonal matrix, $L$ is the lower triangular matrix, $U$ is the upper triangular matrix.

$$
D = \left[\begin{matrix}
    \alpha &&0 &&0\\
    0 &&\alpha &&0\\
    0 &&0 &&\alpha
    \end{matrix}\right] \quad , \quad
L = \left[\begin{matrix}
    0 &&0 &&0\\
    0 &&0 &&0\\
    1 &&0 &&0
    \end{matrix}\right] \quad , \quad
U = \left[\begin{matrix}
    0 &&0 &&1\\
    0 &&0 &&0\\
    0 &&0 &&0
    \end{matrix}\right] 
$$
<br><br>
To solve $Ax=b$, where $x = \left(\begin{matrix}x_1 \\ x_2 \\ x_3 \end{matrix} \right),\space b = \left(\begin{matrix}b_1 \\ b_2 \\ b_3 \end{matrix} \right)$:
<br><br>

# Jacobi method
<br>
$$\begin{align}
(D+L+U)x &= b \\
Dx_{n+1} &= b - (L+U)x_n \\
x_{n+1} &= D^{-1}(b - (L+U)x_n) 
\end{align}$$
<br><br>
Define $e_n = x_n - \bar{x}$, where $A\bar{x}=b$
$$\begin{align}
e_{n+1} = x_{n+1} - \bar{x} &= -D^{-1}(L+U)x_n + D^{-1}b - \bar{x}\\
&= -D^{-1}(L+U)(x_n - \bar{x})
\end{align}$$
<br><br>
Define $G = -D^{-1}(L+U)$
$$\begin{align}
G = -D^{-1}(L+U) &= - \left(\begin{matrix}
    \alpha &&0 &&0\\
    0 &&\alpha &&0\\
    0 &&0 &&\alpha
    \end{matrix}\right)^{-1}
    \left[ \left(\begin{matrix}
    0 &&0 &&0\\
    0 &&0 &&0\\
    1 &&0 &&0
    \end{matrix}\right) +
    \left(\begin{matrix}
    0 &&0 &&1\\
    0 &&0 &&0\\
    0 &&0 &&0
    \end{matrix}\right) \right]\\
&= - \frac{1}{\alpha}
    \left(\begin{matrix}
    0 &&0 &&1\\
    0 &&0 &&0\\
    1 &&0 &&0
    \end{matrix}\right)
\end{align}$$
<br><br>
$$\begin{align}
\therefore e_{n+1} &= Ge_n\\
e_n &= G^ne_0
\end{align}$$
<br><br>
If $\rho(G) < 1$, then $\lim_{n\to\infty}\|G^n\| = 0$.<br>
Then, $\lim_{n\to\infty}\|G^ne_0\| = \lim_{n\to\infty}\|e_n\| = 0 $
<br><br>
If G is diagonally dominant, then $\rho(G)<1$.<br>
But, $$|a_{11}| = 0,\space |a_{12}|+|a_{13}|= \frac{1}{\alpha}\\
|a_{11}| \ngtr |a_{12}|+|a_{13}|$$
Hence, G is not diagonally dominant.
<br><br>
It does not converge to unique solution.
<br><br>


# Gauss-Seidal method
$$\begin{align}
(D+L+U)x &= b\\
(D+L)x_{n+1} &= -Ux_n + b\\
x_{n+1} &= -(D+L)^{-1}Ux_n + (D+L)^{-1}b
\end{align}$$
<br><br>
Define $e_n = x_n - \bar{x}$, where $A\bar{x}=b$
$$\begin{align}
e_{n+1} = x_{n+1} - \bar{x} &= -(D+L)^{-1}Ux_n + (D+L)^{-1}b - \bar{x}\\
&= -(D+L)^{-1}U(x_n - \bar{x})
\end{align}$$
<br><br>
Define $G = -(D+L)^{-1}U$
$$\begin{align}
G = -(D+L)^{-1}U &= - \left(\begin{matrix}
    \alpha &&0 &&0\\
    0 &&\alpha &&0\\
    1 &&0 &&\alpha
    \end{matrix}\right)^{-1}
    \left(\begin{matrix}
    0 &&0 &&1\\
    0 &&0 &&0\\
    0 &&0 &&0
    \end{matrix}\right)\\
&= - \left(\begin{matrix}
    \frac{1}{\alpha} &&0 &&0\\
    0 &&\frac{1}{\alpha} &&0\\
    -\frac{1}{\alpha^2} &&0 &&\frac{1}{\alpha}
    \end{matrix}\right)
    \left(\begin{matrix}
    0 &&0 &&1\\
    0 &&0 &&0\\
    0 &&0 &&0
    \end{matrix}\right)\\
&= - \left(\begin{matrix}
    0 &&0 &&\frac{1}{\alpha}\\
    0 &&0 &&0\\
    0 &&0 &&-\frac{1}{\alpha^2}
    \end{matrix}\right)
\end{align}$$
<br><br>
If $\rho(G) < 1$, then $\lim_{n\to\infty}\|G^n\| = 0$.<br>
Then, $\lim_{n\to\infty}\|G^ne_0\| = \lim_{n\to\infty}\|e_n\| = 0 $
<br><br>
If G is diagonally dominant, then $\rho(G)<1$.<br>
But, $$|a_{11}| = 0,\space |a_{12}|+|a_{13}|= \frac{1}{\alpha}\\
|a_{11}| \ngtr |a_{12}|+|a_{13}|$$
Hence, G is not diagonally dominant.
<br><br>
It does not converge to unique solution.
